```
# Lambda Function Example
func = lambda x: "Positive" if x > 0 else "Negative" if x < 0 else "Zero"
print(func(5))
print(func(-3))
print(func(0))
```

```
# Lambda with List Comprehension
func = [lambda arg=x: arg * 10 for x in range(1, 5)]
for i in func:
    print(i())
```

```
# Lambda Returning Multiple Results
calc = lambda x, y: (x + y, x * y)
res = calc(3, 4)
print(res)
```

```
# Creating Sets
s = {1, 2, 3}
print(s)

s = set([1, 2, 3])
print(s)

s = set()
print(s)

s = {10, 50, 20}
print(s)
print(type(s))

# Typecasting list to set
s = set(["a", "b", "c"])
print(s)

# Duplicate values
s = {"Geeks", "for", "Geeks"}
print(s)
```

```
# Adding and Removing Elements
s = {"a", "b", "c"}
s.add("d")
s.remove("c")
print(s)
```

```
# Union of Sets
a = {"x", "y"}
b = {"y", "z"}
u = a.union(b)
print(a | b)
print(u)
```

```
# Intersection of Sets
a = {1, 2, 3}
b = {2, 3, 4}
i = a.intersection(b)
print(a & b)
print(i)
```

```
# Difference of Sets
a = {1, 2, 3}
b = {2, 3, 4}
d = a.difference(b)
print(a - b)
print(d)
```

```
# Symmetric Difference
a = {1, 2, 3}
b = {2, 3, 4}
print(a ^ b)
```

```
# Clearing a Set
s = {1, 2, 3}
s.clear()
print(s)
```

```
# Subset and Superset
A = {1, 2}
B = {1, 2, 3, 4}
print(A.issubset(B))
print(B.issuperset(A))
```

```
# Membership and Iteration
print(3 in {1, 2, 3})
for item in {1, 2, 3}:
    print(item)
```

```
# Set Comprehension
s = {x*x for x in range(5)}
print(s)
```
