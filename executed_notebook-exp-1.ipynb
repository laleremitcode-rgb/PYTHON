# Types of Operators in Python
This notebook demonstrates different Python operators with examples.

## Arithmetic Operators

```python
x = 10
y = 20

print('x + y =', x + y)
print('x - y =', x - y)
print('x * y =', x * y)
print('x / y =', x / y)
print('x % y =', x % y)
print('x // y =', x // y)
print('x ** y =', x ** y)
```

## Comparison Operators

```python
x = 20
y = 25

print('x == y is', x == y)
print('x != y is', x != y)
print('x > y is', x > y)
print('x < y is', x < y)
print('x >= y is', x >= y)
print('x <= y is', x <= y)
```

## Logical Operators

```python
x = 10
y = 20

print('x and y is', x and y)
print('x or y is', x or y)
print('not x is', not x)
```

## Bitwise Operators

```python
a = 10
b = 4

print(a & b)
print(a | b)
print(a ^ b)
print(~a)
print(a << 2)
print(a >> 2)
```

## Membership Operators

```python
a = 10
b = 7
lst = [1,2,3,4,5,6,7,8,9]

if a in lst:
    print("a is present in the list")
else:
    print("a is not present in the list")

if b in lst:
    print("b is present in the list")
else:
    print("b is not present in the list")

x = 23
y = 20
lst = [10,15,20,30,35,40,45,50]

if x not in lst:
    print("x is not available in the list")
else:
    print("x is available in the list")

if y not in lst:
    print("y is not available in the list")
else:
    print("y is available in the list")
```

## Identity Operators

```python
x = 6

if type(x) is int:
    print("true")
else:
    print("false")

x = 4.5

if type(x) is not int:
    print("true")
else:
    print("false")
```
